```python
import sys
sys.path.append('..')

from virtuous.GenerationModel import GenerationModel
from virtuous.FinancialModel import FinancialModel
```

```python
start_year = 2022         # year the power plant was built
investigate_year = 2026   # year we are simulating for
```

```python
""" simulate energy generation """
generation = GenerationModel(
    start_year=start_year,
    peak_energy=100 * 0.2 * (30 * 24), # 100MW * capacity_factor=0.2 and hours per month
    seasonality=0.5,
    degradation=0.05/12, # 5% per year
    uncertainty=0.1,
    )
monthly_data = generation.calculate_year(investigate_year)
# monthly_data
```

```python
""" calculate costs and revenues for this energy """
financial = FinancialModel(
    start_year=start_year,
    energy_price=25,
    cleaning_cost=30000,
    maintenance_cost=40000,
    )
monthly_data = financial.extend_with_financial(
    monthly_data=monthly_data,
    year=investigate_year)
monthly_data
```
